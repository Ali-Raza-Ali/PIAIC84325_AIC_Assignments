# Part1

```python
us_census = pd.concat(map(pd.read_csv, glob.glob(os.path.join('Assignment/', "states*.csv"))))
```

```python
us_census = us_census.drop('Unnamed: 0',axis=1).reset_index(drop=True)
```

```python
us_census.columns
```

```python
us_census.dtypes
```

```python
us_census.head()
```

```python
menAndwomen = us_census['GenderPop'].str.split('_').dropna()
```

```python
us_census['men'] = np.nan
us_census['women'] = np.nan
```

```python
for i in range(len(menAndwomen)):
    us_census['men'].values[i] = [int(i[0].split('M')[0]) for i in menAndwomen][i]
    if menAndwomen[i][1].split('F')[0] !='':
        us_census['women'].values[i]=menAndwomen[i][1].split('F')[0]
    else:
        us_census['women'].values[i]=np.nan
#     us_census['women'].values[i] = [int( for i in menAndwomen if i =='' return 0 ][i]
```

```python
us_census['Income'] = us_census['Income'].str.split("$").str[1].astype(float)
```

```python
plt.scatter(us_census.women, us_census.Income)
```

```python
us_census['women'] = us_census['TotalPop'] - us_census['men']
```

```python
# us_census = us_census.fillna(method='ffill')
```

```python
print(us_census.duplicated().unique())#True showing duplicate exist
us_census.duplicated()
```

```python
us_census = us_census.drop_duplicates()
```

```python
plt.scatter(us_census.women, us_census.Income)
```

```python
us_census.columns
```

```python
percentCols = ['Hispanic', 'White', 'Black', 'Native', 'Asian',
       'Pacific']
```

```python
for i in percentCols:
    us_census[i] = us_census[i].str.split("%").str[0].astype(float)
```

```python
us_census[percentCols].plot.hist()
plt.show()
```
